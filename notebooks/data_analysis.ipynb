```python
import pandas as pd
import json
```

```python
df = pd.read_csv("../data.csv")
df
```

```python
def parse_loss(loss):
    loss = json.loads(loss)
    loss_vector = loss['Value']
    return loss_vector

steps = json.loads(df['Loss'][0])['Step']
steps
df['Loss Vector'] = df['Loss'].apply(parse_loss)
df
```

```python
df['ID'] = df['Experiment Name'].str.split('__', expand=True)[0]
df['ID'] = df['ID'].str.replace('fashion_mnist_experiment_', '')

df['ID'].astype(int)
```

```python
df_losses = pd.DataFrame(df['Loss Vector'].to_list(), columns=steps)
df_losses.set_index(df['ID']).T
```
